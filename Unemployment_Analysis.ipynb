```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# Create Sample Dataset
# -----------------------------
data = {
    'Month': ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],
    'Unemployment_Rate': [7.8,7.5,8.2,10.5,12.3,11.7,10.9,9.8,8.7,7.9,7.4,7.1]
}

df = pd.DataFrame(data)

# -----------------------------
# Display Dataset
# -----------------------------
print("Dataset:")
print(df)

# -----------------------------
# Basic Information
# -----------------------------
print("\nSummary Statistics:")
print(df.describe())

# -----------------------------
# Visualization 1: Line Plot
# -----------------------------
plt.figure(figsize=(10,5))
plt.plot(df['Month'], df['Unemployment_Rate'], marker='o')
plt.title('Monthly Unemployment Rate')
plt.xlabel('Month')
plt.ylabel('Unemployment Rate (%)')
plt.grid()
plt.show()

# -----------------------------
# Visualization 2: Bar Plot
# -----------------------------
plt.figure(figsize=(10,5))
sns.barplot(x='Month', y='Unemployment_Rate', data=df)
plt.title('Unemployment Rate by Month')
plt.show()

# -----------------------------
# Highest and Lowest Rate
# -----------------------------
highest = df.loc[df['Unemployment_Rate'].idxmax()]
lowest = df.loc[df['Unemployment_Rate'].idxmin()]

print("\nHighest Unemployment:")
print(highest)

print("\nLowest Unemployment:")
print(lowest)
```
